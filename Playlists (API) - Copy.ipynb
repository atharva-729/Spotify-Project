```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
cid = 'client_id'
secret = 'client_secret'
from spotipy.oauth2 import SpotifyOAuth

redirect_uri = "http://localhost:8888/callback"

# Scopes for authorization
scopes = ["playlist-read-private"]  # Include 'playlist-read-private'

# Create SpotifyOAuth object
sp_oauth = SpotifyOAuth(cid, secret, redirect_uri, scope=scopes)

# Open the Spotify authorization URL in the user's web browser
auth_url = sp_oauth.get_authorize_url()
print(f"Open this URL in your browser to grant access: {auth_url}")

# After user authorization, obtain the access token
# This part will involve user interaction (potentially in a separate step)
code = input("Enter the code from the redirect URI: ")
token_info = sp_oauth.get_access_token(code)

# Create a Spotipy object with the access token
sp = spotipy.Spotify(auth=token_info['access_token'])
```
